## Type of Clustering : 
1) Hierarchical Clustering, 2) K-mean Clustering, 3) DBSCAN Clustering 

### K-means clustering

## KNN Flow Chart :
 1) K=3,
 2) Partiton,
 3) Compute Centroid,
 4) Find Distances from all data points to all centroids,
 5) Move data point to Nearest centroid,
 6) Recompute Centroid,
 7) Repeat steps 3, 4 and 5,
 8) If we dont have to move data point to cluster then we stop,

```python
from sklearn.cluster import KMeans
```

```python
SSE = []
for i in range(1,10):
    kmeans = KMeans(n_jobs = -1, n_clusters = i, init='k-means++')
    kmeans.fit(data.iloc[:,[0,1,2,3]])
    SSE.append(kmeans.inertia_)
```

```python
df = pd.DataFrame({'Cluster':range(1,10), 'SSE':SSE})
plt.figure(figsize=(12,6))
plt.plot(df['Cluster'], df['SSE'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia');
plt.title("'ELBOW METHOD TO DETERMINE OPTIMAL VALUE OF 'K'\n");
```

```python
kmeans = KMeans(n_jobs = -1, n_clusters = 3, init='k-means++')
kmeans.fit(data.iloc[:,[0,1,2,3]])
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
data['cluster'] = kmeans.labels_

data
```

```python
display(data['cluster'].value_counts(), data['Species'].value_counts())
```

```python
plt.figure(figsize=(10,5))
plt.scatter(data['SepalLengthCm'], data['SepalWidthCm'], c=data.cluster)
plt.title('Predicted Clusters\n')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1] ,  s=200, c='red', label = 'Centroids')
plt.show()
```
